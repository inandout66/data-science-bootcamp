```python
import matplotlib
matplotlib.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
_ = plt.plot([0, 1], [0, 3]) # warto≈õci na osi X 0 1 i Y 0 3, przypisanie tego do zmiennej tymczasowej _ powoduje brak napisu
```

```python
plt.plot([0, 1], [0, 3], label='line', color='green') #parametr label oznacza opis linii, parametr color kolor
plt.xlabel('o≈õ x') #opis osi X
plt.ylabel('o≈õ y') #opis osi x
plt.title('Wykres liniowy') #tytu≈Ç wykresu
plt.legend() #wy≈õwietla wykres z legendƒÖ
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green') #parametr label oznacza opis linii, parametr color kolor
plt.xlabel('o≈õ x') #opis osi X
plt.ylabel('o≈õ y') #opis osi x
plt.title('Wykres liniowy') #tytu≈Ç wykresu
plt.legend() #wy≈õwietla wykres z legendƒÖ
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x,x_sin)
plt.plot(x,x_cos)
```

Wiele wykres√≥w na raz

```python
plt.subplot(121) #1 wiersz, 2 ilo≈õƒá kolumn, numer wykresu tutaj 1
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122) #1 wiersz, 2 kolumny, numer wykresu tutaj 2
plt.plot([0, 1, 2], [3, 1, 4], 'r--') #r-- czerwona linia przerywana
```

```python
fig = plt.figure(figsize=(12,8))

plt.subplot(131) #1 wiersz, 3 ilo≈õƒá kolumn, numer wykresu tutaj 1
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132) #1 wiersz, 3 kolumny, numer wykresu tutaj 2
plt.plot([0, 1, 2], [3, 1, 4], 'r--') #r-- czerwona linia przerywana
plt.subplot(133) #1 wiersz, 3 kolumny, numer wykresu tutaj 3
plt.plot([0, 1, 2], [3, 1, 4], 'ro') #ro czerwony wykres punktowy
```

Wykresy s≈Çupkowe

```python
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2])
```

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2)
```

```python
x3 = np.random.randn(300) * 50

plt.scatter(x=x1, y=x2, s=x3)
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```
