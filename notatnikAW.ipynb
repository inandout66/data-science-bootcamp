```python
import numpy as np #importuje bibliotekÄ™ numpy
import pandas as pd #importuje bibliotekÄ™ pandas

def fetch_financial_data(company='AMZN'): #funkcja do importu ze stooq
  import pandas_datareader.data as web
  return web.DataReader(company, data_source='stooq')
df = fetch_financial_data('UBER')

df['Average'] = (df.Open + df.Close)/ 2.0 #Å›rednia kursu dziennego

df = df.sort_index() #sortowanie

df['Close_shift'] = df.Close.shift(1) #przesuwam o 1, Å¼eby porÃ³wnywaÄ‡ do danych za dzieÅ„ poprzedni

df['Daily_change'] = df.Close / df.Close_shift - 1 #tworzÄ™ kolumnÄ™ Daily_change - dziennej zmiany

```

```python
df_positive = df[df.Daily_change > 0] #zapisane dane w tablicy df_positive sesje, w ktÃ³rych dzienna zmiana jest >0
df_positive
```

```python
df_positive.Daily_change.mean() #wyliczam Å›redniÄ… z dodatnich wartoÅ›ci zmiany dzienniej
```

```python
df_negative = df[df.Daily_change < 0] #zapisane dane w tablicy df_negative sesje, w ktÃ³rych dzienna zmiana jest <0
df_negative
```

```python
df_negative.Daily_change.mean() #wyliczam Å›redniÄ… z ujemnych wartoÅ›ci zmiany dzienniej
```

```python
import numpy as np #importuje bibliotekÄ™ numpy
import pandas as pd #importuje bibliotekÄ™ pandas

def fetch_financial_data(company): #funkcja do importu ze stooq
  import pandas_datareader.data as web
  return web.DataReader(company, data_source='stooq')
apple = fetch_financial_data('AAPL')
amazon = fetch_financial_data('AMZN')
google = fetch_financial_data('GOOGL')
uber = fetch_financial_data('UBER')
```

```python
apple.columns = ['apple_' + col.lower() for col in apple.columns] #dodanie apple_ przed nazwÄ™ kolumny
amazon.columns = ['amazon_' + col.lower() for col in amazon.columns]
google.columns = ['google_' + col.lower() for col in google.columns]
uber.columns = ['uber_' + col.lower() for col in uber.columns]
```

```python
apple.head()
```

```python
df = pd.concat(objs=[apple, amazon, google, uber], axis= 1) #poÅ‚Ä…czenie danych w jednÄ… tablicÄ™ wzglÄ™dem osi 1
```

```python
pd.set_option('display.float_format', lambda x: f'{x:.2f}') #ustawiony format 2 liczb po przecinku

df.describe().T #describe po transpozycji
```
